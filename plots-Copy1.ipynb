```python
df = pd.read_excel("chips.ods", engine="odf", sheet_name="timeline")
# setting the figure size
width = 11
height = 4
sns.set(rc={'figure.figsize': (width, height)})
# Plotting with Seaborn
sns.set(style="whitegrid")

# Create figure and axis objects
fig, ax1 = plt.subplots(figsize=(11, 5), constrained_layout=True)

# Scatter plot
sns.scatterplot(data=df, x='Year', y='incoming data flow  [byte/year]', style="Experiment at CERN", hue='Experiment at CERN', s=350, ax=ax1, palette="deep")
ax1.set_yscale("log")
ax1.set_xlabel("Year")
ax1.set_ylabel("Data Centre data flow [bytes/year]")

ax1.set_xticks(df['Year'])
ax1.set_xticklabels(df['Year'], rotation=45)

# Secondary y-axis for the plot
ax2 = ax1.twinx()
ax2.plot(df['Year'], df['Cpu Clock cycle'], color='red', label='CPU clock cycle [Hz]')
ax2.set_yscale("log")
ax2.set_ylabel("CPU clock cycle [Hz]")
ax2.yaxis.set_label_position("right")

# Show legend for both axes
ax1.legend(loc='upper right', bbox_to_anchor=(1, 2))
ax2.legend(loc='lower right')
sns.despine()

plt.grid(False)
# plt.savefig('images/networking/network_timeline_1.jpg', dpi=600, bbox_inches='tight')
plt.show()
plt.close()
```
